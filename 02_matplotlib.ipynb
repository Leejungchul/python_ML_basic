```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xarr = np.random.rand(15).cumsum()  # cumsum : 누적합

yarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(4,3))
plt.plot(xarr)
plt.show()
```

```python
plt.figure()
plt.subplot()  # axes(축)이 만들어짐
plt.show()
```

```python
plt.figure()
plt.subplot() 
plt.plot(xarr)
plt.show()
```

```python
# 그래프 여러개 그리기 1
plt.figure()
plt.subplot(1,2,1) # (1,2,1) -> 순서 : 1
plt.plot(xarr)
plt.subplot(1,2,2) 
plt.plot(xarr)
plt.show()
```

```python
# 그래프 여러개 그리기 2
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(xarr)
ax2 = fig.add_subplot(1,2,2)
ax2.plot(xarr)
plt.show()
```

```python
# 순서가 중요함
plt.figure()
plt.subplot(1,2,1)
plt.subplot(1,2,2) 
plt.plot(yarr)
plt.plot(xarr)
plt.show()
```

```python
# 그래프 여러개 그리기 3
fig, axes = plt.subplots(2,2) 
axes[0,0].plot(xarr)
axes[1,1].plot(yarr)
plt.show()
```

```python
fig, axes = plt.subplots(2,2) 
axes = axes.flatten()

for i in range(4):
    axes[i].plot(xarr)
plt.show()
```

```python
xarr = np.random.rand(15).cumsum()  # cumsum : 누적합
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(8,6))
plt.plot(xarr, label='korean', color="m", linestyle='--', marker='o')
plt.plot(yarr,label='english', color='gray',linestyle='-.',marker='^')
plt.plot(zarr,label='japanese', color='gold',linestyle=':', marker='*')
plt.title("this is title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.show() 
```

#### 산점도 scatter plot

```python
plt.scatter(xarr, yarr, c='r', s = 100)
plt.scatter(xarr,zarr)
plt.show()
```

### Barplot

```python
year = ['A','B','C']
points = [100,200,300]

plt.bar(year, points)
plt.show()
```

```python
year = ['A','B','C']
points = [100,200,300]

plt.barh(year, points)
plt.show()
```

### Boxplot

```python
x = np.random.normal(0,2.0,100)
y = np.random.normal(-3,1.5,500)
z = np.random.normal(1.2,1.5,1000)
```

```python
plt.boxplot([x,y,z])
plt.show()
```
