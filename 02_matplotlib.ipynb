# matplotlib
데이터 시각화를 위한 파이썬의 시각화 툴

```python
# matplotlib 임포트 하기
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 생성
# rand : 표준정규분포에서 15개 난수 생성
# cumsum : 누적합
xarr = np.random.rand(15).cumsum()  
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

### 그래프
![그래프](https://nextjournal.com/data/QmT9YMFX4JAyn7Rznu9VV8SKhwkDdifgDkpcULqL8Ws1ur?filename=99_matplotlib_structure-1.png&content-type=image/png)

```python
plt.figure(figsize=(8,6))  # figure 크기 설정
plt.plot(xarr, label='korean')
plt.plot(yarr,label='english')
plt.plot(zarr,label='japanese')
plt.title("this is title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(xarr, label='korean', color="m", linestyle='--', marker='o')
plt.plot(yarr,label='english', color='gray',linestyle='-.',marker='^')
plt.plot(zarr,label='japanese', color='gold',linestyle=':', marker='*')
plt.title("this is title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.show() 
```

### scatter plot 산점도

```python
plt.scatter(xarr, yarr)
plt.show()
```

```python
# 설정 c : 색, s : 크기
plt.scatter(xarr, yarr, c='r', s = 100)
plt.scatter(xarr,zarr, marker='^')
plt.show()
```

### Bar plot


```python
year = ['A','B','C']
points = [100,200,300]

plt.bar(year, points, color=['m','coral','gold'])
plt.show()
```

```python
year = ['A','B','C']
points = [100,200,300]

plt.barh(year, points, color=['m','coral','gold'])
plt.show()
```

### 박스 플롯

```python
x = np.random.normal(0,2.0,100)
y = np.random.normal(-3,1.5,500)
z = np.random.normal(1.2,1.5,1000)
```

```python
plt.boxplot([x,y,z])
plt.show()
```
