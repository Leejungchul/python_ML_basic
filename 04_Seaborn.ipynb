```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
titanic = pd.read_csv('./dataset/titanic.csv')
titanic.head()
```

### histogram

```python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='Age')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='Age', bins=8)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='Age', bins=8, hue='Survived')
plt.show()
```

### distplot


```python
sns.displot(data=titanic, x='Age')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', col='Survived', hue='Sex')
plt.show()
```

---
### 막대 그래프
- x축이 범주형 데이터고 y축에 개수로 표현할 수 있는 데이터에 사용


```python
titanic['Pclass'].value_counts()
```

```python
sns.countplot(data=titanic, x='Pclass')  # countplot 자체가 value_counts()를 해주는 기능이 존재함
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data = titanic)   # pclass별 fare를 비율로 표시
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data = titanic, ci=False) 
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data = titanic, ci=False, hue='Sex')
plt.show()
```

---
### point plot
- x축을 범주형으로 함

```python
sns.pointplot(x='Pclass', y = 'Fare', hue='Sex', data=titanic)
plt.show()
```

---
### box plot
- categorical 데이터기 때문에 x축은 범주형 데이터 사용

```python
sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=titanic, hue='Sex')
plt.show()
```

---
### violin plot

```python
sns.violinplot(x='Pclass',y='Age', data=titanic)
plt.show()
```

```python
sns.violinplot(x='Pclass',y='Age', data=titanic, hue='Sex', split=True)
plt.show()
```

---
### countplot

```python
sns.countplot(x='Embarked', data=titanic)
plt.show()
```

```python
sns.countplot(x='Sex', data=titanic)
plt.show()
```

```python
sns.countplot(x='Survived', data=titanic)
plt.show()
```

### scatter plot

```python
tips = sns.load_dataset('tips')
type(tips)
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex',data=tips)
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)   # regplot : 선형회귀선을 그려주는 그래프

plt.show()
```
